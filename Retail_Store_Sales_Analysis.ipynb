```python
data = data.drop(columns=['Customer ID', 'Item'], errors='ignore')
data
```

```python
#create Month and Year Colums
#Extract Month and Year from Tranaction Date
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'])
data['Month'] = data['Transaction Date'].dt.strftime('%B')
data['Year'] = data['Transaction Date'].dt.year
data
```

```python
#Rename Column names
data.rename(columns={
    "Price Per Unit":"Price_Per_unit",
    "Total Spent":"Total_Spent",
     "Location":"Purchase_Mode"
},inplace=True)
data
```

```python
#fill with mean
data["Price_Per_unit"].fillna(data["Price_Per_unit"].mean(),inplace=True)
data
```

```python
#fill with 0
data["Quantity"].fillna(0,inplace=True)
data
```

```python
#fill with 0
data["Total_Spent"].fillna(0,inplace=True)
data
```

```python
data['Discount Applied'].dtype
data['Discount Applied'].unique()
```

```python
data["Discount Applied"].fillna("unknown",inplace=True)
data
```

```python
#visualisation
1.#Bar chart
import matplotlib.pyplot as plt
data["Total_Spent"]=data["Price_Per_unit"]*data["Quantity"]
Total_Spent=data.groupby("Category")["Total_Spent"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.bar(Total_Spent.index,Total_Spent.values,color="green")
plt.title("Top 10 Ttems by Total Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

 **Interpretation**
* "Butchers" has the highest total sales,meaning meat products are the top selling caregory.
* "Electric Household essentials" and "Beverages" also have high sales,indicating strong demand for these items.
* Categories like "Patisserie" and "Milk Products" have relatively lower total sales but still make it into the top 10.
* The difference between categories is not extremely large,suggesting that sales are somewhat evently distributed among the top items,but a few(like butchers)lead slightly



```python
2.#Scatterplot
import matplotlib.pyplot as plt
#randomly Select 10 rows
sample_data=data.sample(10)
plt.figure(figsize=(10,6))
plt.scatter(sample_data["Category"],sample_data["Total_Spent"])
plt.xlabel("Category")
plt.ylabel("Total_Spent")
plt.title("Category vs Total Spent(random 10 rows)")
plt.xticks(rotation=45)
plt.show()

```



### ðŸ§© Interpretation:

* The chart shows that spending amounts vary widely  across categories â€” some transactions have very high spending (for example, Furniture and Milk Products), while others have low spending  (*Butchers* and *Electric household essentials*).

* The scatter points show that each purchase amount differs  even within the same category â€” meaning customers spend different amounts on items from the same category.




```python
3.#piechart
data["Payment Method"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    colors=["Blue","Orange","green"])
plt.title("Payment Method distribution in retail store sales")
plt.ylabel("")
plt.show()

```

### ðŸ§© **Interpretation:**
* **Cash** payments make up the **largest share (34.3%)**, meaning slightly more customers prefer to pay with cash.
* **Digital Wallets** are also very popular, accounting for **33.0%** of total sales â€” almost equal to cash payments.
* **Credit Cards** are used in **32.8%** of sales, which is only a little lower than the other two methods.


```python
4.#Box Plot
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.boxplot(x='Payment Method',y='Price_Per_unit', data=data)
plt.title('Box Plot of Price Per Unit by Payment Method', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

### ðŸ“Š **Interpretation**
* The **median (middle line)** of all three payment methods is almost the same (around 23â€“24), meaning the typical price per unit is similar regardless of how the payment is made.
* The **interquartile range (IQR)** â€” the box portion â€” is also similar, suggesting that price variation within each payment method is quite consistent.
* The **minimum and maximum values (whiskers)** extend roughly from 5 to 40 for all methods, indicating a similar overall range of prices.
* There are **no visible outliers**, meaning most price values fall within a normal range.




```python
5.#barchart
data_2024=data[data["Year"]==2024]
data['Month'] = pd.Categorical(data['Month'],
    categories=['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December'],
    ordered=True
)
Month_totals=data_2024.groupby("Month")["Total_Spent"].sum().sort_index()
plt.figure(figsize=(10,6))
plt.bar(Month_totals.index,Month_totals.values,color="red")
plt.xlabel("Month")
plt.ylabel("Total_Spent")
plt.title("Total Spent by Month in 2024")
plt.xticks(rotation=45)
plt.show()
```

### ðŸ“Š **Interpretation**

* The **highest total spending** occurred in **December** and **January**, both close to â‚¹48,000â€“â‚¹49,000.
  â†’ This could indicate higher shopping or festive season expenses.
* The **lowest total spending** is seen in **February** (around â‚¹37,000), showing a dip in spending after the New Year period.
* Other months like **April**, **June**, and **November** also show moderately high spending (â‚¹44,000â€“â‚¹46,000).
* Overall, spending remains fairly consistent across the months, with only slight fluctuations.




```python
6.#Heat Map
data['Month'] = pd.Categorical(data['Month'],
    categories=['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December'],
    ordered=True
)
# Create a pivot table: Total Spent by Category and Month
heatmap_data = data.pivot_table(
    index='Category',
    columns='Month',
    values='Total_Spent',
    aggfunc='sum'
)
# Plot the heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, annot=True, fmt="0.1f",cmap="RdYlBu",linewidths=1.5, linecolor='white')
plt.title("Total Spent by Category and Month")
plt.xlabel("Month")
plt.ylabel("Category")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

### ðŸ§  **Interpretation:**

* **January** shows the **highest spending** across most categories â€” possibly due to new-year purchases.
* **Computers and electric accessories** have very **high spending in January** (â‰ˆ 25,188.5) but **low in October** (â‰ˆ 11,431.8).
 * **Butchers** and **Food** categories also start strong in January but gradually **decrease over the months**.
* **Furniture** spending spikes in **April** (â‰ˆ 19,808.0) â€” maybe due to seasonal sales or furnishing trends.
* **Overall spending** tends to **decline slightly toward the end of the year**, with **Octoberâ€“December** showing lower values for most categories.




```python
#7.Bubble chart
category_counts = data['Category'].value_counts().reset_index()
category_counts.columns = ['Category', 'Count']
min_size = 100
max_size = 2000
sizes = (
    (category_counts['Count'] - category_counts['Count'].min()) /
    (category_counts['Count'].max() - category_counts['Count'].min())
) * (max_size - min_size) + min_size
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    x=category_counts['Category'],
    y=category_counts['Count'],
    s=sizes,
    color='skyblue',
    alpha=0.7,
    edgecolors='k'
)
plt.title('Bubble Chart: Category Counts', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## ðŸ§  **Interpretation:**
* **Electric household essentials** and **Furniture** have the **largest bubbles**, showing they had the **highest number of transactions**.
* **Food** and **Milk Products** also have **high transaction counts**, just slightly below the top categories.
* **Beverages** and **Butchers** fall in the **middle range**, with moderate transaction volumes.
* **Computers and electric accessories** show a **lower number of transactions**.
* **Patisserie** has the **smallest bubble**, indicating it has the **least number of transactions** among all categories.




```python
8.#Donut Chart
import plotly.express as px
data_random = data.sample(n=5, random_state=42)
data_agg = data_random.groupby('Category')['Total_Spent'].sum().reset_index()
fig = px.pie(
    data_agg,
    names='Category',
    values='Total_Spent',
    hole=0.4,
    title='Random Sample: Total Sales by Category'
)
fig.show()
```

```python
9.#Line chart
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'])
data_2023 = data[data['Transaction Date'].dt.year == 2023]
data_2023['Month'] = data_2023['Transaction Date'].dt.month_name()
monthly_spent = data_2023.groupby('Month')['Total_Spent'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])
plt.figure(figsize=(10, 6))
plt.plot(monthly_spent.index, monthly_spent.values, color='green', marker='o', linewidth=2)
plt.title('Monthly Variation of Total Spent in 2023', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Total_Spent')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Interpretation:

* The **highest total spending** occurred in **January**, indicating strong sales or higher expenses at the beginning of the year.
* Spending **dropped sharply in February** and reached its **lowest point in March**.
* From **April to July**, total spending **gradually increased**, peaking again in **July**.
* After July, spending **declined in August**, then showed **fluctuations** â€” rising in **September**, dipping again in **October and November**, and **rising slightly in December**.




```python
10.#violin chart
plt.figure(figsize=(10,6))
sns.violinplot(data=data, x='Purchase_Mode', y='Total_Spent', inner='quartile', palette='pastel')
plt.title('Distribution of Total Spent by Purchase Mode')
plt.xlabel('Purchase_Mode')
plt.ylabel('Total_Spent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### **Interpretation:**

* Both **Online** and **In-store** purchases have a **similar distribution pattern**, meaning spending behavior is fairly comparable across modes.
* The **widest part** of each violin indicates where most transactions are concentrated â€” around **lower to mid spending levels (roughly between 50 and 200)**.
* A few **higher spending values (above 300)** appear in both modes, but these are less frequent (the upper thin tails).
* The **median line** (center dashed line) is roughly similar for both purchase modes, suggesting **no major difference in median total spending** between online and in-store purchases.




```python
11.#Staked barchart
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'])
data_2022 = data[data['Transaction Date'].dt.year == 2022]
category_discount_2022 = pd.crosstab(
    data_2022['Category'],
    data_2022['Discount Applied']
)
category_discount_2022.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    color=['red','orange','green']
)
plt.title(' Category vs Discount Applied (2022)', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.legend(title='Discount Applied')
plt.tight_layout()
plt.show()
```

### **Interpretation:**
* Each bar represents a product category, divided into **three parts**:
    * **Red** (no discount applied),
    * **Orange** (discount applied), and
    * **Green** (unknown discount status).
* **Butchers** have the **highest total transactions**, followed by **Electric household essentials** and **Milk Products**.
* **Food** and **Computers & Electric Accessories** show comparatively **fewer transactions**.
* In all categories, transactions are **fairly balanced** among discount types, but **discounted and unknown** transactions together make up a **large portion** of sales.
